```python
def load_word2vec():
    """ Load Word2Vec Vectors
        Return:
            wv_from_bin: All 3 million embeddings, each lengh 300
    """
    import gensim.downloader as api
    wv_from_bin = api.load("word2vec-google-news-300")
    vocab = list(wv_from_bin.vocab.keys())
    print("Loaded vocab size %i" % len(vocab))
    return wv_from_bin
```

```python
wv_from_bin = load_word2vec()
```

```python
for words in antonyms:
    a_d = wv_from_bin.distance(words[0], words[1])
    guesses1 = wv_from_bin.most_similar(words[0])
    for g in guesses1:
        if g[1] > 0.5 and (g[1] + 0.1) < a_d:
            print(words, a_d, "_____", words[0], g)
    guesses2 = wv_from_bin.most_similar(words[1])
    for g in guesses2:
        if g[1] > 0.5 and (g[1] + 0.1) < a_d:
            print(words, a_d, "_____", words[1], g)

w1 = "poor"
w2 = "rich"
w3 = "wealthy"
# guesses = wv_from_bin.most_similar(positive=['poor', 'rich'], negative=['good'])
# print(guesses)
w1_w2_dist = wv_from_bin.distance(w1, w2)
w1_w3_dist = wv_from_bin.distance(w1, w3)
print("Synonyms {}, {} have cosine distance: {}".format(w1, w2, w1_w2_dist))
print("Antonyms {}, {} have cosine distance: {}".format(w1, w3, w1_w3_dist))
```

```python
for v in vocab:
    guesses = wv_from_bin.most_similar(v)
    for g in guesses:
        for gg in guesses:
            if g[1] > 0.5 and gg[1] > 0.5 and g[0] != gg[0]:
                sim = wv_from_bin.similarity(g[0], gg[0])
                if sim < 0.1:
                    print(v, ":", g, gg,  "____", sim)
```
